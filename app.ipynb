```python
import pandas as pd
```

```python
import glob

# Encuentra todos los archivos xlsx en inputs/ordenes_new
archivos_new = glob.glob("inputs/ordenes_new_v2/*.xlsx")

# Lee todos los archivos y los concatena en un solo DataFrame
df_new = pd.concat([
    pd.read_excel(
        archivo,
        dtype={
            "N° Orden": str,
            "Año Orden": str,
            "Código Empresa": str
        }
    )
    for archivo in archivos_new
], ignore_index=True)

df_new
```

```python
df_old = pd.read_excel(
    "inputs/ordenes_old.xlsx",
    dtype={
        "ORDEN": str,
        "AÑO": str,
        "REGIMEN": str,
        "ADUANA": str,
        "CODIGO DE CLIENTE": str
    }
)
df_old
```

```python
df = df_new.merge(df_old, left_on=["N° Orden", "Año Orden", "Código Empresa"], right_on=["ORDEN", "AÑO", "CODIGO DE CLIENTE"], how="outer")
df
```

```python
df = df[(df["CODIGO DE CLIENTE"].isin(df_new["Código Empresa"]) | df["N° Orden"].notna())
        & ~(df["Año Orden"].isin(["2030", "2040"]))
        ]
```

```python
df["numer"] = (df["F. Numeración"].isna() & df["FECHA NUMERACION"].isna()) | (df["F. Numeración"].notna() & df["FECHA NUMERACION"].notna())
df["llegada"] = (df["Fecha de llegada"].isna() & df["FECHA DE LLEGADA"].isna()) | (df["Fecha de llegada"].notna() & df["FECHA DE LLEGADA"].notna())
df["retiro"] = (df["Fecha Retiro"].isna() & df["FECHA DE RETIRO"].isna()) | (df["Fecha Retiro"].notna() & df["FECHA DE RETIRO"].notna())
df["levante"] = (df["Fecha Levante"].isna() & df["FECHA DE LEVANTE"].isna()) | (df["Fecha Levante"].notna() & df["FECHA DE LEVANTE"].notna())
df["eta"] = (df["ETA_x"].isna() & df["ETA_y"].isna()) | (df["ETA_x"].notna() & df["ETA_y"].notna())
df["fact"] = (df["Fecha Facturación"].isna() & df["FECHA DE CIERRE DE FACTURACION"].isna()) | (df["Fecha Facturación"].notna() & df["FECHA DE CIERRE DE FACTURACION"].notna())

df["numer"].value_counts(), df["llegada"].value_counts(), df["retiro"].value_counts(), df["levante"].value_counts(), df["eta"].value_counts(), df["fact"].value_counts()
```

```python
df["no_cruzan"] = df["N° Orden"].isna() | df["ORDEN"].isna()
df["no_cruzan"].value_counts()
```

```python
df["rev_numer"] = (df["numer"]==False) & (df["ORDEN"].notna()) & (df["FECHA NUMERACION"].notna())
df["rev_llegada"] = (df["llegada"]==False) & (df["ORDEN"].notna()) & (df["FECHA DE LLEGADA"].notna())
df["rev_retiro"] = (df["retiro"]==False) & (df["ORDEN"].notna()) & (df["FECHA DE RETIRO"].notna())
df["rev_levante"] = (df["levante"]==False) & (df["ORDEN"].notna()) & (df["FECHA DE LEVANTE"].notna())
df["rev_eta"] = (df["eta"]==False) & (df["ORDEN"].notna()) & (df["ETA_y"].notna())
df["rev_fact"] = (df["fact"]==False) & (df["ORDEN"].notna()) & (df["FECHA DE CIERRE DE FACTURACION"].notna())
df["rev_numer"].value_counts(), df["rev_llegada"].value_counts(), df["rev_retiro"].value_counts(), df["rev_levante"].value_counts(), df["rev_eta"].value_counts(), df["rev_fact"].value_counts()
```

```python
# Asegurarse de que 'Fecha de apertura' es datetime
df["Fecha de apertura"] = pd.to_datetime(df["Fecha de apertura"], errors="coerce")

fecha_corte = pd.Timestamp("2026-05-01")

df["rev_business"] = (df["Fecha de apertura"] < fecha_corte) & (df["Estado actual"] != "Facturada")

# Para los registros donde rev_business es True, poner todas las columnas 'rev_' (excepto 'rev_business')
# y también numer, llegada, retiro, levante, eta, fact en False
extra_false_cols = ["numer", "llegada", "retiro", "levante", "eta", "fact"]
rev_columns = [col for col in df.columns if col.startswith("rev_") and col != "rev_business"]
all_set_false = rev_columns + [col for col in extra_false_cols if col in df.columns]
df.loc[df["rev_business"], all_set_false] = False

df["rev_business"].value_counts()
```

```python
df_rev_business = df[df["rev_business"]]
df_rev_business.drop(columns=["rev_business", "rev_numer", "rev_llegada", "rev_retiro", "rev_levante", "rev_eta", "rev_fact", "numer", "llegada", "retiro", "levante", "eta", "fact", "no_cruzan"], inplace=True)
df_rev_business
```

```python
with pd.ExcelWriter("outputs/ordenes_cruzadas.xlsx") as writer:
    df.to_excel(writer, sheet_name="Total", index=False)
    df_rev_business.to_excel(writer, sheet_name="Negocio", index=False)
```
